```python
import sympy as sm
import numpy as np

#Schemat Dormanda-Prince'a
def DP(f, a, b, x0, N):
  h = (b-a)/N
  T = [a]
  X = [x0]
  t = a
  x = x0
  for i in range(N):
    D1 = h*f(t, x)
    D2 = h*f(t + h/5, x + D1/5)
    D3 = h*f(t + 3*h/10, x + 3*D1/40 + 9*D2/40)
    D4 = h*f(t + 4*h/5, x + 44*D1/45 - 56*D2/15 + 32*D3/9)
    D5 = h*f(t + 8*h/9, x + 19372*D1/6561 - 25360*D2/2187 + 64448*D3/6561 - 212*D4/729)
    D6 = h*f(t + h, x + 9017*D1/3168 - 355*D2/33 + 46732*D3/5247 + 49*D4/176 - 5103*D5/18656)
    D7 = h*f(t + h, x + 35*D1/384 + 500*D3/1113 + 125*D4/192 - 2187*D5/6784 + 11*D6/84)
    x = x + 35*D1/384 + 500*D3/1113 + 125*D4/192 - 2187*D5/6784 + 11*D6/84
    t = t + h
    T.append(t)
    X.append(x)
  return T, X

#Sprawdzenie dokładności schematu Dormanda-Prince'a dla różnych równań
def Test1(N):
  Err = [] #Inicjalizacja tabeli zawierającej błędy dla poszczególnych funkcji
  T1, X1 = DP(f1, 0, 5, 1, N) #Wartości dla x'=x, x(0)=1 na [0, 5]
  Err.append(abs(X1[-1]-np.exp(5))) #Porównanie z wartością w 5 dla rozwiązania dokładnego exp(t)
  T2, X2 = DP(f2, -np.pi/4, 0, 0, N) #Wartości dla x'=x^2+1, x(-pi/4)=0 na [-pi/4, 0]
  Err.append(abs(X2[-1]-1)) #Porównanie z wartością w 0 dla rozwiązania dokładnego tg(t+pi/4)
  T3, X3 = DP(f3, 0, 5, 1, N) #Wartości dla x'=-x, x(0)=1 na [0, 5]
  Err.append(abs(X3[-1]-np.exp(-5))) #Porównanie z wartością w 5 dla rozwiązania dokładnego exp(-t)
  T4, X4 = DP(f4, 0, 5, 1, N) #Wartości dla x'=2*x*t, x(0)=1 na [0, 5]
  Err.append(abs(X4[-1]-np.exp(25))) #Porównanie z wartością w 5 dla rozwiązania dokładnego exp(t^2)
  return Err

#Sprawdzenie dokładności i rzędu schematu Dormanda-Prince'a dla równania x'=x, x(0)=1 przy zwiększeniu liczby kroków o 2 w każdej iteracji
def Test2():
  Err = [] #Inicjalizacja tabeli zawierającej błędy dla algorytmów o krokach o różnej długości
  P = [] #Inicjalizacja tabeli zawierającej prybliżone wartości rzędu dla algorytmów o krokach o różnej długości
  N=100 #początkowa ilość kroków
  for i in range(4):
    T1, X1 = DP(f1, 0, 5, 1, N)
    Err.append(abs(X1[-1]-np.exp(5)))
    if i == 0:
      p = 0
    else:
      p = np.log2(Err[i-1]/Err[i]) #prybliżenie rzędu za pomocą faktora, o który błąd jest zmniejszany
    P.append(p)
    N = N*2 #zwiększenie liczby kroków o 2
  return Err, P

#Funkcja opisująca równanie x'=x
def f1(t, x):
  return x

#Funkcja opisująca równanie x'=x^2+1
def f2(t, x):
  return x**2 + 1

#Funkcja opisująca równanie x'=-x
def f3(t, x):
  return -x

#Funkcja opisująca równanie x'=2*t*x
def f4(t, x):
  return 2*t*x

Err1 = Test1(6000)
Err2, P2 = Test2()
print("Błąd dla różnych funkcji dla 6000 iteracji:")
for i in range(4):
  print(f"f{i+1}: {Err1[i]:.4e}")
print("Błąd i rząd rozwiązania dla  kroków o rónych wielkościach:")
print("Długość kroku (h)\tBłąd\t\tRząd")
for i in range(4):
    print(f"{5/(100*(i+1)):.6f}\t\t{Err2[i]:.4e}\t{P2[i]:.4f}")
```
